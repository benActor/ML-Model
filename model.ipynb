# 3. replace Income by 0 and 1

```python
binary_values = {'<=50K': 1, '>50K': 0}

final_df = data_set.replace({"income": binary_values})

```

```python
# here it is appropriate to manually encode educational rather than using the processor
eductional_levels ={'1st-4th': 1, 
                    '5th-6th':2,
                    '7th-8th':3,
                    '9th': 4,
                    '10th': 5, 
                    '11th':6,
                    '12th': 7,
                    'Some-college': 8, 
                    'Preschool': 9,
                    'HS-grad': 10,
                    'Prof-school': 10,
                    'Assoc-acdm':11,
                    'Assoc-voc': 11,
                    'Bachelors' : 12,
                    'Masters':13,
                    'Doctorate': 14,
                     }


final_df = final_df.replace({"education": eductional_levels})
```

# 4. Separate the dataset into train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_df.drop('income', axis = 1)

y = final_df['income']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)


print('train data size {}'.format(X_train.shape[0]))

print('test data size {}'.format(X_test.shape[0]))
```

# 5. Process Data 

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer


#transformer = ColumnTransformer(transformers=[('cat', OneHotEncoder(), [0, 1])])



categorical_features = ["marital.status", "relationship" , "race", "sex"]

num_features = ["age", "workclass", "occupation", "capital.gain", "capital.loss", "hours.per.week"]


categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])



numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])


preprocessor = ColumnTransformer(sparse_threshold=0,
    transformers=[
        ('num', numeric_transformer, num_features),
        ('cat', categorical_transformer, categorical_features)], remainder='passthrough')
```

```python
X_train_processed = preprocessor.fit_transform(X_train)

X_test_processed = preprocessor.fit_transform(X_test)
```

```python
X_train_processed.shape

```

# 6. Perform K-NN  Classification

```python
from sklearn.neighbors import KNeighborsClassifier


# we start with a hyper parameter of 3
n = 3 

classifier = KNeighborsClassifier(n)
```

```python
classifier.fit(X_train_processed, y_train)
```

```python
predictions = classifier.predict(X_test_processed)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix


confusion = confusion_matrix(y_test, predictions)


confusion
```

# 7. Perform SVM

```python
from sklearn.svm import SVC

clf = SVC(kernel="linear", C=1)

clf.fit(X_train_processed, y_train)

preds = clf.predict(X_test_processed)
```

```python
print('Train set accuracy: {:.2f}'.format(clf.score(X_train_processed, y_train)))
print('Test set accuracy: {:.2f}'.format(clf.score(X_test_processed, y_test)))
```

# 8. Perform Linear Discriminant Analysis classifier

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda_clf = LinearDiscriminantAnalysis()

lda_clf.fit(X_train_processed, y_train)

lda_preds = lda_clf.predict(X_test_processed)
```

```python
print('Train set accuracy: {:.2f}'.format(lda_clf.score(X_train_processed, y_train)))
print('Test set accuracy: {:.2f}'.format(lda_clf.score(X_test_processed, y_test)))
print('\n')

lda_confusion = confusion_matrix(y_test, lda_preds)

print('Confusion matrix:\n', lda_confusion)
```

# 9. Logistic Regression classifier

```python
from sklearn.linear_model import LogisticRegression

lrc_clf = LogisticRegression(C=1)

lrc_clf.fit(X_train_processed, y_train)

lrc_preds = lrc_clf.predict(X_test_processed)

print('Train set accuracy: {:.2f}'.format(lrc_clf.score(X_train_processed, y_train)))
print('Test set accuracy: {:.2f}'.format(lrc_clf.score(X_test_processed, y_test)))
print('\n')

lrc_confusion = confusion_matrix(y_test, lrc_preds)

print('Confusion matrix:\n', lrc_confusion)
```
